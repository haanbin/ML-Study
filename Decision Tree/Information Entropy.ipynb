# What is meaning of information?

- Low Probability Event : High information(Surprising)
- High Probability Event : Low information(Unsurprising)

$$information(x) = - log(p(x))$$

- Why base-2 logarithm?
- Why choose negative value?

## import library

```python
from math import log2
from matplotlib import pyplot
import numpy as np
```

```python
# probability of the event
p = 0.5

# calculate information for event
h = -log2(p)

# print result
print('p(x) : {}, information :  {} bits'.format(p,h))
```

```python
from math import log2
# probability of the event
p = 0.4

# calculate information for event
h = -log2(p)

# print result
print('p(x) : {}, information :  {} bits'.format(p,h))
```

## Relationship between probability & Information Gain

- High Probability --> Low Information Gain

```python
probs = np.arange(0.1,1.1,0.1)
info = [-log2(i) for i in probs]

pyplot.figure(figsize=(20,10))
pyplot.plot(probs, info, marker = '.')
pyplot.title('probabilty vs Information')
pyplot.xlabel('probability')
pyplot.ylabel('Information')
```

## Calculate the Entropy for a Random Variable

- What is Random Variable?
- Simply, calculating the information for a random variable is called "information Entroy"

- condition : Random Variable X with k in N discrete states

$$H(X) = - \sum_{k=1}^N p_i \times \log_{2} p_i$$

```python
# the number of events
n = 6

# probability of one event
p = 1.0 / n

# calculate entropy
entropy = - sum([p* log2(p) for _ in range(n)])

print('entropy : %.3f bits' % entropy)
```

```python
def cal_entropy(n):
    p = 1.0 / n
    entropy = - sum([p * log2(p) for _ in range(n)])
    return entropy

# Single Information Gain VS Entropy
cal_entropy(2)
```

# What if Skewed Probability Distribution?

- Skewed Probability Distribution(Unsurprising) : Low Entropy
- Balanced Probability Distribution(Surprising) : High Entropy

```python

# Define probabilities 
probs = np.arange(0.0,0.6,0.1)

# Create probability distribution
dists = [[p, 1.0 - p] for p in probs]

# Calculate entropy 
def entropy(events, ets = 1e-10):
    return -sum([p * log2(p+ets) for p in events])

# Calculate entropy for each distribution
ents = [entropy(d) for d in dists]

# Plot probability distribution VS Entropy
pyplot.figure(figsize = (20,10))
pyplot.title('probability Distribution VS Entropy')
pyplot.plot(ents)
#pyplot.xticks(probs)
pyplot.xlabel('probability Distribution')
pyplot.ylabel('Entropy (bits)')
```
